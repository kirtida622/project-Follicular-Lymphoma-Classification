```python
# libs
import numpy as np
import pandas as pd

import tensorflow as tf
from tensorflow.keras import layers, models
from skimage.feature import greycomatrix, greycoprops

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, cohen_kappa_score

import os
import pickle
from PIL import Image
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

import glob
import os
import numpy as np
from skimage.feature import greycomatrix, greycoprops
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.preprocessing import StandardScaler

CNN_MODEL = os.path.join('..', 'Models_part2', 'cnn_model.pkl')
ANN_MODEL = os.path.join('..', 'Models_part2', 'ann_model.pkl')
```

```python
def load_image(image_path, image_size=(128, 128)):
    
    images = []
    try:
        # Open and preprocess the image
        img = Image.open(image_path).convert('RGB')  # Convert to RGB
        img = img.resize(image_size)  # Resize image to target size
        images.append(np.array(img))  # Convert image to numpy array
    except Exception as e:
        print(f"Error loading image {file_path}: {e}")

    # Convert lists to numpy arrays
    images = np.array(images, dtype='float32') / 255.0  # Normalize pixel values to [0, 1]

    return images

def save_model(model, file_path):
    """
    Saves the Model to a .pkl file.
    """
    with open(file_path, 'wb') as file:
        pickle.dump(model, file)
        
    print(f"Model saved to {file_path}")
    
def load_model(file_path):
    """
    Loads the model from a .pkl file. 
    """
    with open(file_path, 'rb') as file:
        model = pickle.load(file)
        
    print(f"Model loaded from {file_path}")
    
    return model

# Function to calculate GLCM properties
def calculate_glcm_features(image, distances, angles):
    properties = ['contrast', 'dissimilarity', 'homogeneity', 
                  'energy', 'correlation', 'ASM']
    glcm = greycomatrix(image, 
                        distances = distances, 
                        angles = angles,
#                         levels = 256,
                        symmetric = True, 
                        normed = True)
    feature_vector = []
    
    for prop in properties:
        prop_values = greycoprops(glcm, prop)
        feature_vector.extend(prop_values.flatten())
    
    return feature_vector

# Fetch CNN features, GLCM features
def extract_feature(image, cnn_model, distances, angles):
    for i, img in enumerate(image):
#         print(f"Processing image {i+1}/{len(images)}")
        
        # CNN Features
        image_expanded = np.expand_dims(img, axis=0)  # Expand dimensions for the CNN model
        cnn_features = cnn_model.predict(image_expanded).flatten()

        # GLCM Features
        image_gray = np.mean(img, axis=-1).astype(np.uint8)  # Convert to grayscale if needed
        glcm_features = calculate_glcm_features(image_gray, distances, angles)

        # Combine CNN + GLCM features
        combined_features = np.concatenate([cnn_features, glcm_features])

    return combined_features
```

```python
# Load image
image_path = os.path.join('..', 'Dataset', 'Malignant_Lymphoma_Classification', 'FL', 'sj-05-588-R1_005.tif')
image = load_image(image_path, image_size=(224, 224))
print(f"Loaded {len(image)} images.")
print(f"Image of shape {image.shape}")

# Load CNN Model for feature extraction
cnn_model = load_model(CNN_MODEL)
print(f"Loaded CNN_MODEL {cnn_model}.")

# Feature extraction
distances, angles = [1], [0]
image_feature = extract_feature(image, cnn_model, distances, angles)
print(f"image_feature : {image_feature} \nlen : {len(image_feature)} ")

# Load ANN model
ann_model = load_model(ANN_MODEL)
print(f"Loaded ANN_MODEL {ann_model}.")

# prediction
y_predict = np.argmax(ann_model.predict(np.array(image_feature).reshape(1, -1)), axis=1) # np.argmax(model.predict(X_test), axis=1) 
print(f"The predicted class for the image is: {y_predict[0]}")

    
```
